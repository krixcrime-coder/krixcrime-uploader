## PART 1 — Dashboard Password Hash banao

```
# Cell 1: bcrypt install karo
!pip install bcrypt -q
```

```
import bcrypt

# APNA PASSWORD YAHAN LIKHO (jo dashboard login ke liye use karoge)
MY_PASSWORD = "apna-password-yahan"  # <-- Ye change karo

hashed = bcrypt.hashpw(MY_PASSWORD.encode(), bcrypt.gensalt()).decode()

print("="*60)
print("DASHBOARD_PASSWORD_HASH (Vercel mein paste karo):")
print("="*60)
print(hashed)
print("="*60)
print(f"\nYe hash aapke password ke liye hai: '{MY_PASSWORD}'")
print("Password yaad rakhna — hash se wapas nahi dekh sakte!")
```

## PART 2 — YouTube Refresh Token lao

**Pehle Google Cloud Console mein:**
1. [console.cloud.google.com](https://console.cloud.google.com) kholo
2. APIs & Services → Credentials → apna OAuth 2.0 Client ID kholo
3. **Authorized redirect URIs** mein add karo: `http://localhost`
4. Client ID aur Client Secret niche paste karo

```
# Cell 3: Library install karo
!pip install google-auth-oauthlib -q
```

```
# Cell 4: Apna Client ID aur Secret daalo
CLIENT_ID     = "xxxxxxxx.apps.googleusercontent.com"  # <-- Google Cloud se copy karo
CLIENT_SECRET = "GOCSPX-xxxxxxxxxxxxxxxxxx"             # <-- Google Cloud se copy karo

from google_auth_oauthlib.flow import InstalledAppFlow

client_config = {
    "installed": {
        "client_id": CLIENT_ID,
        "client_secret": CLIENT_SECRET,
        "auth_uri": "https://accounts.google.com/o/oauth2/auth",
        "token_uri": "https://oauth2.googleapis.com/token",
        "redirect_uris": ["urn:ietf:wg:oauth:2.0:oob", "http://localhost"]
    }
}

SCOPES = ["https://www.googleapis.com/auth/youtube.upload"]

flow = InstalledAppFlow.from_client_config(client_config, SCOPES)

# Colab ke liye console flow use karo
flow.redirect_uri = "urn:ietf:wg:oauth:2.0:oob"
auth_url, _ = flow.authorization_url(prompt="consent", access_type="offline")

print("="*60)
print("STEP 1: Niche diya link kholo (YouTube wala Gmail account se login karo):")
print("="*60)
print(auth_url)
print("="*60)
print("\nLogin ke baad ek CODE milega — usse niche wali cell mein paste karo")
```

```
# Cell 5: Wahan se mila CODE yahan paste karo
auth_code = input("Google se mila CODE yahan paste karo: ")

flow.fetch_token(code=auth_code)
creds = flow.credentials

print("\n" + "="*60)
print("YT_REFRESH_TOKEN (GitHub Secret mein save karo):")
print("="*60)
print(creds.refresh_token)
print("="*60)
print("\n✅ Ye token permanent hai — baar baar generate karne ki zaroorat nahi!")
print("⚠️  Kisi ke saath share mat karna — YouTube channel ka full access hai isme")
```
